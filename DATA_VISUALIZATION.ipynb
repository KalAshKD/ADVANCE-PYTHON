# Data Visualisation 

# Scatter Plot / Correlation Plot
## used when we have to find correlation between two numeical variables

```python
import pandas as pd 
import numpy as np
import matplotlib.pyplot as plt 
```

```python
toyota_file = pd.read_csv("Toyota.csv",index_col=0,na_values=["??","????"])
toyota_file.dropna(axis=0,inplace=True)
print(toyota_file.head())
```

## plt.scatter(x/df['Col_name'],y/df['Col_name'],c='color')

```python
plt.grid()
plt.scatter(toyota_file['Age'],toyota_file['Price'],c='Blue')
plt.title('AGE vs PRICE')
plt.xlabel('Age')
plt.ylabel('PRICE')
plt.show()

# Histogram
## used to represent frequency distribution of numerical variable
```

## plt.hist(df_name['col_name']/x,color='color_name',edgecolor='color_name',bins=n)

```python
plt.hist(toyota_file['KM'],color='Green',edgecolor='White',bins=5)
plt.title("KM histo")
plt.xlabel("KM---------->")
plt.ylabel("Frequency-------->")
plt.show()
```

# BarPlot
## Used to represent the  frequency distribution of categoial variable 


## plt.bar(x,height of bars/y,color='')

## plt.xtics(location,labels,rotation=90)

```python
counts = [979,120,12]
fueltype = ('Petrol',"Diesel","CNG")
index = np.arange(len(fueltype))

plt.bar(index,counts,color=["Red","Green","Blue"])
plt.xticks(index,fueltype,rotation = 90)
plt.xlabel("fueltype")
plt.ylabel("Frequency")
plt.show()
```

# RegPlot(Scatter Plot through Seaborn Library)
## sns.regplot(x=df['col_name'],y=df['col_name],fit_reg=True/False,marker="*")


```python
import seaborn as sns
```

```python
sns.set(style="whitegrid") #dargrid,dark,whitegrid,white
sns.regplot(x=toyota_file['Age'],y=toyota_file['Price'],fit_reg=True,marker="*")
plt.title("regPlotPLot")
```

# lmplot
## sns.lmplot(x='col_name',y='col_name',data=datframe_name,fit_reg=True/False,
## hue='3rdcol_name',legend=True,paleette="Set1")

Here fit_reg adds or plots the line in the graph ,hue is 3rd variable, legend means notation, 

```python
sns.set()
sns.lmplot(x='Age',y='Price',data=toyota_file,fit_reg=False,hue="FuelType",legend=True,palette="Set1")
plt.title("LMPLot")
```

# distplot(Histogram from sns)
## sns.distplot(x/dataframe_name['col_name'],kde=True/False,bins = n,color=color_name)
here kde means kernel density estimate which gives curves for frequency distribution 


```python
sns.set(style="whitegrid")
sns.distplot(toyota_file['Age'],kde=True,bins=5,color="Red")
plt.title("DistPLot")
```

# CountPlot(Bar Plot through sns)
## sns.count(x='col_name',data=df_name,,palette="Set1")

```python
sns.set()
sns.countplot(x="FuelType",data=toyota_file,palette="Set1")
```

# Grouped Bar Plot
