## Binary Opeartion

## 1,2,3 )numpy.bitwise_and / or /xor(arr1,arr2)

## numpy.binary_repr(num / array,width)-> use to convert element of array in its binary form


```python
import numpy as np 
```

```python
arr=np.array([[1,2,3],[4,5,6]])
Arr=np.array([[4,5,6],[7,8,9]])
Bitwise_And=np.bitwise_and(arr,Arr)
print(Bitwise_And)
print

Bitwise_Or=np.bitwise_or(arr,Arr)
print(Bitwise_Or)
print

Bitwise_XOR=np.bitwise_xor(arr,Arr)
print(Bitwise_XOR)
print

# representation of numbers into binary 
A = np.binary_repr(5,width=4)
print(A)
B = np.binary_repr(8,width=8)
print(B)
```

# 2) MATHEMATICAL OPERATION

## 2.1) Trigonometric Operation

```python
import matplotlib.pyplot as plt
```

```python
# sine
""" syntax:->m numpy.sin(x[,out])=ufunc'sin)"""

arr=np.array([0,np.math.pi/2,1/2,3*np.math.pi/2,2*np.math.pi])
sinx=np.sin(arr)
print(sinx)

plt.plot(sinx)
plt.show()

#same for cos tan arcsin,arccos,arctan,sinh,cosh,arcsinh,arccosh,arctanh
```

```python
arr2=np.array([0,np.math.pi/2,1/2,3*np.math.pi/2,2*np.math.pi])
sinx=np.sin(arr2)
print(sinx)

plt.plot(sinx,marker="D",mec="aqua",color="black",lw=3)
plt.show()
```

## 2.2) Rounding Operation
Around(arr,decimals)

```python
Arr=np.linspace(0,10,10,dtype=float)
print(Arr)
Round_Arr = np.around(Arr,decimals=2)
print(Round_Arr)
```

## 2.3) Exponents and Logirithmic Functions 

```python
exponential = np.exp(Arr)
print(exponential);print

log_10 = np.log10(Arr)
print(log_10);print

log_2 = np.log2(Arr)
print(log_2)

```

## 2.5) Arthimatic

```python
np.reciprocal(Arr)
```

```python
print np.floor_divide(Arr,Arr)

print np.conj(Arr)

print np.cbrt(Arr)
print np.clip(Arr,a_min=0,a_max=10)
```
