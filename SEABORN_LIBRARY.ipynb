## Histogram using Seaborn with Kernel Density Estimate

```python
import numpy as np 
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
Data = pd.read_csv('ToyotaCorolla.csv',index_col=0,na_values=["??","????"])
sns.distplot(Data['KM'])
```

## Histogram using Seaborn without Kernel Density Estimate

```python
sns.set(style="darkgrid")
sns.distplot(Data['KM'],kde=False)
```

Set bins=5

```python
sns.set(style="darkgrid")
sns.distplot(Data['KM'],kde=False,bins=5,color="red")
```
